# 10.1 Matplot의 기초

## 직선 그래프

```python
%matplotlib inline
import matplotlib.pyplot as plt
y = [1,2,3,4,5]
plt.plot(y)
plt.show()
```

### plot() 메소드의 입력 파라미터의 개수가 2개인 경우
- 첫 번째 항목은 x축 좌표값
- 두 번째 항목은 y축 좌표값

```python
y = [13, 16, 15, 18, 16, 17, 16]
plt.plot(range(len(y)), y);    #세미콜론을 넣으면 2D라인 문자열 생성 X
```

### x축에 의미 있는 문자(열) 넣기

```python
x = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
y = [13, 16, 15, 18, 16, 17, 16]
plt.plot(x, y);
```

### 라벨(label) 입력
- xlabel() : x축의 라벨을 입력
- ylabel() : y축의 라벨을 입력

```python
x = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
y = [13, 16, 15, 18, 16, 17, 16]
plt.plot(x, y);
plt.xlabel('Daily')
plt.ylabel('No. of Books Sold');
```

### 그래프 1개 더 추가

```python
x = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
y1 = [13, 16, 15, 18, 16, 17, 16]
y2 = [17, 14, 17, 16, 15, 15, 14]
plt.plot(x, y1);
plt.plot(x, y2);
plt.xlabel('Daily')
plt.ylabel('No. of Books Sold')
```

### 레전드(legend) 붙이기

```python
x = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
y1 = [13, 16, 15, 18, 16, 17, 16]
y2 = [17, 14, 17, 16, 15, 15, 14]
plt.plot(x, y1, label='Sold');
plt.plot(x, y2, label='On Shelves');
plt.xlabel('Daily')
plt.ylabel('No. of Books Sold')
plt.legend(loc="upper left");
```

### 제목 넣기

```python
x = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
y1 = [13, 16, 15, 18, 16, 17, 16]
y2 = [17, 14, 17, 16, 15, 15, 14]
plt.plot(x, y1, label='Sold');
plt.plot(x, y2, label='On Shelves');
plt.xlabel('Daily')
plt.ylabel('No. of Books Sold')
plt.legend(loc="upper left")
plt.title('Le Petit Prince: Sold & Left');
```

## 점선 그래프

### 'bo' 옵션
- b(Blue) 색상의 'o' 모양 점 표시

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x, y, 'bo')
plt.show()
```

### axis() 메소드
- x축과 y축 범위 지정
 - [xmin, xmax, ymin, ymax]

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x, y, 'bo')
plt.axis([0, 6, 0, 30])
plt.show()
```

|서식|메소드|
|:---:|:---:|
|legend 추가|legend()|
|축 교정|axis()|
|축의 라벨 설정|xlabel(), ylabel()|
|그림 저장|savefig()|

# 10.2 Numpy를 이용한 plot 기능 

```python
import numpy as np
points = np.array([[1,1],[1,2],[1,3],[2,1],[2,2],[2,3],[3,1],[3,2],[3,3]])
p = np.array([2.5, 2])
import matplotlib.pyplot as plt
plt.plot(points[:,0], points[:,1], "ro");
plt.plot(p[0], p[1], "bo");
print(points[:,0])
print(points[:,1])
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 5, 10)
y = x**2
plt.plot(x, y);
```

```python
x = np.linspace(0, 10, 20)
y = x**2.0
plt.plot(x, y, "bo-", linewidth=3, markersize =5);
```

```python
plt.plot(x, y, "gs-", linewidth=1, markersize=3)
```

```python
x = np.linspace(0, 10, 20)
y1 = x**2.0
y2 = x**1.5
plt.plot(x, y1, "bo-", linewidth=2, markersize=12, label="First")
plt.plot(x, y2, "gs-", linewidth=2, markersize=12, label="Second")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc="upper left")
plt.savefig("mplot.pdf")
```

```python
x = np.logspace(-1, 1, 20)
y1 = x**2.0
y2 = x**1.5
plt.plot(x, y1, "bo-", linewidth=2, markersize=12, label="First")
plt.plot(x, y2, "gs-", linewidth=2, markersize=12, label="Second")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc="upper left")
plt.savefig("mplot.pdf")
```

## 표준정규분포 그래프
### standard_normal() 사용


```python
import matplotlib.pyplot
import numpy as np
x = np.random.standard_normal(size=1000)
plt.hist(x)   #결과에 대한 빈도수
```

```python
import matplotlib.pyplot
import numpy as np
x = np.random.standard_normal(size=1000)
plt.hist(x, density=True)   # 빈도수가 아닌 확률로 나타내기
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randn(1000)
plt.hist(x)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.rand(30)
y = np.random.rand(30)
colors = np.random.rand(30)
shape = np.pi * (np.random.rand(30)*20) **2
plt.scatter(x, y, s=shape, c=colors, marker='*', alpha=0.7)
plt.show()
```
