```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='CP949')
data = csv.reader(f)
next(data)
x = []
y = []

max_year = 0
for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '06' and row[0].split('-')[2] == '21' :
            y.append(float(row[-1]))
            x.append(int(row[0].split('-')[0]))
            if max(y) == float(row[-1]) :
                max_year = int(row[0].split('-')[0])
            
y = list(map(float, y))
x = list(map(float, x))

dic = { y:x for x, y in zip(x, y) }


plt.text(dic[max(y)],max(y),
         str(max_year) + ' : ' + str(max(y)),
         color='r',
         horizontalalignment='center',
         verticalalignment='bottom')

plt.plot(x, y)

plt.show()
```

```python
#2
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='CP949')
data = csv.reader(f)
next(data)
high = [] # 
low = [] # 
dis = [] # 
year = [] # 
max_dis = 0
max_year = 0

for row in data :
    if row[-1] != '' :
        if int(row[0].split('-')[0]) >= 2001 and row[0].split('-')[1] == '01' and row[0].split('-')[2] == '05' :
            high.append(float(row[-1]))
            low.append(float(row[-2]))
            dis.append(float(row[-1]) - float(row[-2]))
            year.append(row[0].split('-')[0])
            if max(dis) == float(row[-1]) - float(row[-2]) :
                max_year = row[0].split('-')[0]
                
max_dis = max(dis)

high = list(map(float, high))
low = list(map(float, low))
dis = list(map(float, dis))
year = list(map(float, year))

dic = { dis:year for year, dis in zip(year, dis) }

plt.text(dic[max(dis)],max(dis),
         str(max_year) + ' : ' + str(max(dis)),
         color='r',
         horizontalalignment='center',
         verticalalignment='bottom')

plt.plot(year, high)
plt.plot(year, low)
plt.plot(year, dis)
plt.show()
```
